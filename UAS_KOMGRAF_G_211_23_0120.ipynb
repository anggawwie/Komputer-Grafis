**Import Library**

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_excel('stock.xlsx')
print(df)
```

**2a. Distribusi data pada setiap atribut Purchase Price, Current Price dan Gain/Loss dengan
grafik histogram**

```python
plt.figure(figsize=(18, 20))

# Plot histogram untuk Purchase Price
plt.subplot(3, 1, 1)
plt.hist(df['Purchase Price'], bins=10, color='skyblue', edgecolor='black')
plt.title('Distribution of Purchase Price')
plt.xlabel('Purchase Price')
plt.ylabel('Frequency')

# Plot histogram untuk Current Price
plt.subplot(3, 1, 2)
plt.hist(df['Current Price'], bins=10, color='lightgreen', edgecolor='black')
plt.title('Distribution of Current Price')
plt.xlabel('Current Price')
plt.ylabel('Frequency')

# Plot histogram untuk Gain/Loss
plt.subplot(3, 1, 3)
plt.hist(df['Gain/Loss'], bins=10, color='lightcoral', edgecolor='black')
plt.title('Distribution of Gain/Loss')
plt.xlabel('Gain/Loss')
plt.ylabel('Frequency')

# Mengatur layout agar tidak tumpang tindih
plt.tight_layout()

# Tampilkan plot
plt.show()
```

**2b. Atribut Stock Name dan Symbol yang memiliki Dividen / Share 5 tertinggi**

```python
df_urut= df.sort_values('Dividend/share', ascending=False)
top5dividen = df_urut.head(5)[['Stock Name', 'Symbol', 'Dividend/share']]
print(top5dividen)
```

**2c. Atribut Stock Name dan Symbol yang memiliki Market Value 5 tertinggi**

```python
df_urut = df.sort_values('Market Value', ascending=False)
top5market = df_urut.head(5)[['Stock Name', 'Symbol', 'Market Value']]
print(top5market)
```

**2d. Atribut Stock Name dan Symbol yang memiliki Annual Yield 7 tertinggi**

```python
df_urut = df.sort_values('Annual Yield', ascending=False)
top5annual = df_urut.head(7)[['Stock Name', 'Symbol', 'Annual Yield']]
print(top5annual)
```
